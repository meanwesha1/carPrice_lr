```python
#We will train a linear regression model on a car sales dataset and try to predict the prices of the cars.
```

```python
import pandas as pd
import numpy as np
cars = pd.read_csv('imports-85.data')
pd.options.display.max_columns = 99
print(cars.head(3))
```

```python
#Column names are missing in the dataframe, so we will add them.
cols = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'wheel-base', \
        'length', 'width', 'height', 'curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', \
        'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cars.columns = cols
```

```python
#Also, we could see some question marks in the 2nd column for missing data.
#We will replace them with nan.
```

```python
cars['normalized-losses'].unique()
cars = cars.replace('?', np.nan)
print(cars.head(3))
```

```python
# We will look for null values.
print(cars.isnull().sum())
```

```python
#We will remove whichever row is having null value for 'price' column, because we are going to train the model for that only.
```

```python
cars = cars.dropna(subset=['price'])
print(cars.isnull().sum())
```

```python
print(cars.describe(include='all'))
```

```python
#We see some continuous variables like 'price' are stored as object. We'll clean that first.
cat_vars = ['symboling', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'engine-type', 'num-of-cylinders', \
            'fuel-system']
cars[cat_vars] = cars[cat_vars].astype('category')
cont_vars = ['normalized-losses','wheel-base','length','width','height','curb-weight','engine-size','bore','stroke', 'compression-rate','horsepower','peak-rpm',\
             'city-mpg','highway-mpg','price']
cars[cont_vars] = cars[cont_vars].astype('float')
```

```python
#We will fill the null values in the continuous variables with their mean values.
x = cars.isnull().sum()
df_missingvals = cars[x[(x>0)].index]
missing_floats = df_missingvals.select_dtypes(include='float')
cars[missing_floats.columns] = missing_floats.fillna(missing_floats.mean())
```

```python
print(cars.isnull().sum())
```

```python
print(cars['num-of-doors'].value_counts())
```

```python
#Fill the null values in 'num-of-doors' with the most frequent value 'four'.
cars['num-of-doors'] = cars['num-of-doors'].fillna('four')
```

```python
print(cars[cat_vars].describe())
```

```python
#Linear regression assumes that the features are linearly correlated with the target column.
#But for a categorical feature like engine-location a change in its value from front to back or vice-versa has no correlation with the target variable.
#So,instead of having a single column with n integer codes, we create n binary columns out of it.
```

```python
dummy_df = pd.get_dummies(cars[['symboling','make','body-style','drive-wheels','engine-type','num-of-cylinders','fuel-system']],dtype='int64')
dummy_df.head()
cars = pd.concat([cars, dummy_df], axis=1)
cars = cars.drop(['symboling','make','body-style','drive-wheels','engine-type','num-of-cylinders','fuel-system'], axis=1)
```

```python
#We will calculate the correlation between features and target and select only those features having more than 0.2 correlation cofficient with the the target.
corrmat = cars.corr()
price_corrs = corrmat['price'].abs().sort_values()
selected_cols = price_corrs[price_corrs>0.2].index
cars2 = cars[selected_cols].copy()
print(cars2.shape[1])
```

```python
#We will look for collinearity within the columns.
#Since, is difficult to visually compare the values of a 30x30 matrix, we will select only those columns having values>0.7.
#But before that, we will convert the correlation matrix into a non-symmetric one to avoid considering the same relation more than once.
```

```python
corrmat2 = cars2.corr()
new_corrmat2 = corrmat2.iloc[:-1,:-1]
for i in range(new_corrmat2.shape[1]):
    for j in range(i,new_corrmat2.shape[1]):
        new_corrmat2.iloc[i,j]=0
import seaborn as sns
import matplotlib.pyplot as plt        
fig,axes = plt.subplots(figsize=(20,10)) 
sns.heatmap(new_corrmat2.abs())
plt.show()        
```

```python
def highcorr(series):
    x = series.abs()>0.7
    if(x.any()):
        return True
    return False
z = new_corrmat2.apply(highcorr)
highcorr_cols = z[z==True].index
fig,axes = plt.subplots(figsize=(15,8))
sns.heatmap(new_corrmat2[highcorr_cols].abs(),annot=True)
plt.show()
```

```python
# Considering 0.7 as the threshold, we can see that..
# wheel-base is highly correlated with length, width, curb-weight.
# drive-wheels_fwd is highly correlated with drive-wheels_rwd.
# city-mpg is highly correlated with highway-mpg, horsepower, curb-weight.
# length is highly correlated with width, curb-weight.
# highway-mpg is highly correlated with horsepower, curb-weight.
# width is highly correlated with curb-weight, engine-size.
# horsepower is highly correlated with curb-weight, engine-size.
# curb-weight is highly correlated with engine-size.
```

```python
#Scaling the dataframe values.
minimum = cars2.min(axis=0)
maximum = cars2.max(axis=0)
temp = (cars2 - minimum)/(maximum - minimum)
#getting variance of each column.
variance = temp.var()
print(variance)
```

```python
#After comparing the varinces of the correlated columns, we are removing the ones having lower variance than the others.
cars3 = cars2.drop(labels=['wheel-base', 'length', 'width', 'drive-wheels_rwd', 'city-mpg', 'horsepower', 'highway-mpg', \
                           'width', 'engine-size'],axis=1)
print(cars3.shape[1])
```

```python
#Correlation matrix of the feature columns finally selected for training the model.
fig,axes = plt.subplots(figsize=(20,15))
corrmat3 = cars3.corr()
sns.heatmap(corrmat3.abs())#, annot=True)
axes.set_xticklabels(corrmat3.columns, rotation=10)
axes.set_yticklabels(corrmat3.columns, rotation=0)
plt.show()
```

```python
#Preparing the training and the test dataset.
train_cols = corrmat3.columns.drop('price')
shuffled_index = np.random.permutation(cars3.index)
df = cars3.reindex(shuffled_index)
train_size = int(0.75*df.shape[0])
train_df = df.iloc[:train_size].copy()
test_df = df.iloc[train_size:].copy()
```

```python
#Training the model on the dataset.
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train_df[train_cols], train_df['price'])
#Using the model to make predictions.
test_df['predicted_price'] = lr.predict(test_df[train_cols])
#Calculating root-mean-squared error of the model.
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(test_df['price'],test_df['predicted_price'])
rmse = mse**0.5
print(rmse)
```

```python
#The root-mean-squared error value of the model is 2455.43.
#It means that the actual price values of the cars will be within 2455 above or below the price values predicted by the model.
```
